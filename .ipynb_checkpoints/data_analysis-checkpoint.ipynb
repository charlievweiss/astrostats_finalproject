## Load data
Load data file and images from processed_images folder. Add images to dataframe.

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.color import rgb2gray
```

```python
### load from data file

path = "processed_images/"
name = "data.txt"
filename = path + name

# Original data (no images removed to reduce confusion (see next cell))
og_data = pd.read_csv(filename, delimiter = ",", engine='python')

images = []
img_nums = og_data.loc[:,'img_num']

### load images in grayscale
for i in range(0, len(og_data)):
    img_num = img_nums[i]
    img_file = "{}.png".format(img_num)
    img_name = path+img_file

    img = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)
    images.append(img)
    
# add images to dataframe for ease later
og_data['image'] = pd.Series(images, index=og_data.index)

### CHECK:
#img = data['image'].iloc[0]
#plt.imshow(img)
#print(data)
#print(labels)
#print(len(images))
```

### Remove white images
Nearly completely white images confuse the classification algorithms later. This removes them.

```python
color_threshold = 200 # filters anything with a mean above 200 from image (0-255)

data = og_data.copy()

indeces = []
index = 0
for image in data['image']:
    if np.mean(image) > color_threshold:
        indeces.append(index)
    index = index + 1
    
data = data.drop(indeces)
```

## Create training and testing sets

This isn't necessary for unsupervised learning

```python
### split into training and testing sets
num_samples = data.shape[0]
training_size = .99 # percent of data in set (very large now for PCA later -- will generate more data for ML)
testing_size = 1 - training_size

# generate random sample from data for training set, and make testing set all else
msk = np.random.rand(num_samples) < training_size
training_data = data[msk]
testing_data = data[~msk]

### CHECK
# print(training_data)
# print(testing_data)
```

## Pre-process data with Principle Component Analysis
Normally PCA is applied to reduce the dimensionality of data to decrease processing time when applying an algorithm. Since the images are so small, you don't need to do this here. This is merely for visualization purposes, because it's much easier to plot 2 dimensions than 36.

Note: Already have images and data from earlier. A training set isn't necessary for this step.

Resources:
- https://jakevdp.github.io/PythonDataScienceHandbook/05.09-principal-component-analysis.html

```python
# imports
from sklearn.decomposition import PCA
```

```python
### grab images from dataframe:
images = np.array(data.loc[:,'image'])

### Turn each image into vector for PCA
# 2d array: Array of image vectors
images_compressed = []

for i,image in enumerate(images):
    temp = np.asarray(image.flatten()) # turns 2D image array into 1D array
    images_compressed.append(temp)
        
images_compressed = np.asarray(images_compressed)

### CHECK
#test = images[0]
#print(test.shape)
#print(images_compressed[0].shape)
print(images_compressed.shape)
```

```python
### apply PCA

pca = PCA(n_components=2) # n_components = 2 means taking the first 2 directions of variance -- this represents the most representative information
pca.fit(images_compressed)

### Reconstruct faces with principle components (for later figure)
components = pca.transform(images_compressed)
images_projected = pca.inverse_transform(components)

### CHECK
print(images_compressed.shape)
#print(pca.components_)
#print(pca.explained_variance_)
```

## Interesting plots from PCA

```python
### Plot first two principal commponets of each point to learn about the data
# 0 means not a cosmic ray, 1 means it is

## Get amount each image is represented by each component
plot_projected = pca.fit_transform(images_compressed)
## Get corresponding labels from dataframe
labels = np.asarray(data.loc[:,'label'])

## Plot
plt.scatter(plot_projected[:, 0], plot_projected[:, 1], c=labels,cmap='bwr');
plt.title("Images plotted with first 2 principle components")
plt.xlabel('component 1')
plt.ylabel('component 2')

# Legend
cbar = plt.colorbar(ticks=[0,1]);
cbar.ax.set_yticklabels(['not cosmic ray','cosmic ray']);
```

```python
### Visualize Eigen Cosmic Rays (these are what the first 2 most important components look like)
fig, axes = plt.subplots(1,2, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(6,6), cmap='gray')

fig.suptitle("Eigen Cosmic Rays (First 2 Components)");
```

```python
### Plot comparison between some original and reconstructions
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(images[i], cmap='binary_r')
    ax[1, i].imshow(images_projected[i].reshape(6,6), cmap='gray')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('2-dim\nreconstruction');

fig.suptitle("Original vs Reconstructed Images");
```

```python
### See how much info each projection preserves
# This is more informative if you apply PCA with max amount of components

pca = PCA(n_components=36)
pca.fit(images_compressed)

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('percent of image retained');

plt.title("Increase in accuracy from 1 (min) to 36 (max) principle components");
```

### Clustering and Accuracy Score

```python
# clustering
# result: 2 clusters in 36 dimensions. Each cluster center is a 36-dimensional point

# 2 clusters, run 50 times with best result kept
kmeans = KMeans(n_clusters=2, random_state=0, n_init=50)

classifications = kmeans.fit_predict(images_compressed)
kmeans.cluster_centers_.shape
labels = np.array(data.loc[:,'label'])
# find accuracy
acc = accuracy_score(labels, classifications)
print("Accuracy score: {}".format(acc))
```

```python
# Find best and worst fit of both labels

dists = kmeans.transform(images_compressed)
d1 = dists[:,0] # distance from first cluster center
d2 = dists[:,1] # distance from second cluster center

min_d1, min_d2 = np.argmin(d1), np.argmin(d2)

class1 = classifications[min_d1]
class2 = classifications[min_d2]
## CHECK: 
# print(class1, class2)

d1_mean = np.mean(d1)
d2_mean = np.mean(d2)

# filter out incorrect labels from distances
for i in range(0, len(classifications)):
    label = classifications[i]
    # If first label, make point insignificant in second label distances
    if label == class1:
        d2[i] = d2_mean
    # If second label, make point insignificant in first label distances
    elif label == class2:
        d1[i] = d1_mean
#     else:
#         print("this shouldn't happen")
        
mind1, mind2 = np.argmin(d1), np.argmin(d2)
maxd1, maxd2 = np.argmax(d1), np.argmax(d2)
indeces = [[mind1, mind2], [maxd1, maxd2]]
minmax_labels = [[labels[mind1], labels[mind2]], [labels[maxd1], labels[maxd2]]]

# CHECK that min and max values aren't equal to means, and indeces are different
# print(class1_mean, class2_mean)
# print(indeces)
# print(d1[mind1], d2[mind2])
# print(d1[min_d1], d2[min_d2]) # should be the same as first two in row before

print("Cluster 0: {} {}".format(classifications[mind1], classifications[maxd1]))
print("Cluster 1: {} {}".format(classifications[mind2], classifications[maxd2]))
print("True Label C0: {} {}".format(labels[mind1], labels[maxd1]))
print("True Label C1: {} {}".format(labels[mind2], labels[maxd2]))

## Plot images closest and farthest from centroids
fig, ax = plt.subplots(2, 2, figsize=(3, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(2):
    ax[0, i].imshow(images[indeces[0][i]], cmap='binary_r')
    ax[1, i].imshow(images[indeces[1][i]], cmap='binary_r')
    
ax[0, 0].set_ylabel('closest\nin label')
ax[1, 0].set_ylabel('farthest\nin label');

# fig.suptitle("          Cosmic Ray or Not Cosmic Ray?");
fig.suptitle("  Cluster 0 or Cluster 1?");
```

### Training and Testing Sets

```python
## new training and testing sets
### split into training and testing sets
num_samples = data.shape[0]
training_size = .5 # percent of data in set (very large now for PCA later -- will generate more data for ML)
testing_size = 1 - training_size

# generate random sample from data for training set, and make testing set all else
msk = np.random.rand(num_samples) < training_size
training_data = data[msk]
testing_data = data[~msk]

training_images = np.array(training_data.loc[:,'image'])
training_labels = np.array(training_data.loc[:,'label'])
testing_images = np.array(testing_data.loc[:,'image'])
testing_labels = np.array(testing_data.loc[:,'label'])

### CHECK
# print(training_data)
# print(testing_data)
# print(len(training_data))
# print(len(testing_data))
```

```python
def gradient_detection(frame):
    lower_lim = 0
    upper_lim = 255

    laplacian = cv2.Laplacian(frame,cv2.CV_64F, ksize=1) # gradient edge detection in all directions
    sobelx = cv2.Sobel(frame,cv2.CV_64F,1,0,ksize=3) # edge detection in x direction
    sobely = cv2.Sobel(frame,cv2.CV_64F,0,1,ksize=3) # edge detection in y direction

    return laplacian, sobelx, sobely
```

```python
index = 0
normal_img = training_images[index]
normal_label = training_labels[index]
laplacian, sobelx, sobely = gradient_detection(normal_img)

# CHECK: Get insight into values of different edge detections
# print(normal_label)
# print(normal_img)
# print(laplacian)
# print(sobelx)
# print(sobely)

fig, ax = plt.subplots(2,2)
plt.gray()

plt.subplot(2,2,1),plt.imshow(normal_img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
def get_median(img):
    image = img.flatten() #np.asarray(image.flatten())
    image.sort()
#     print(image)
    if (len(image) % 2) == 0: # even
        index = int(len(image)/2)
        median = (image[index-1] + image[index]) / 2
    elif (len(image) % 2) == 1:  # odd
        index = int((len(image) - 1) / 2)
        median = image[index]
    return median
```

```python
# Split training data in cosmic rays and not cosmic rays
CR_inds = []
NCR_inds = []
for i in range(len(training_data)):
    if training_labels[i] == 1:
        CR_inds.append(i)
    elif training_labels[i] == 0:
        NCR_inds.append(i)
    else:
        print("This shouldn't happen")
```

```python
from scipy import stats

"""
Checked:
-mean
-mode
-min
-max
"""

CR_edges = [] # cosmic rays
NCR_edges = [] # not cosmic rays

for i in range(0,len(CR_inds)):
    ind = CR_inds[i]
    image = training_images[ind]
    laplacian, sobelx, sobely = gradient_detection(image)
    max_val = np.max(abs(laplacian))
    min_val = np.min(abs(laplacian))
    median_val = get_median(abs(laplacian))
    mean_val = np.mean(abs(laplacian))
    vals = [max_val, min_val, median_val, mean_val]
    CR_edges.append(vals)
    
for i in range(0,len(NCR_inds)):
    ind = NCR_inds[i]
    image = training_images[ind]
    laplacian, sobelx, sobely = gradient_detection(image)
    max_val = np.max(abs(laplacian))
    min_val = np.min(abs(laplacian))
    median_val = get_median(abs(laplacian))
    mean_val = np.mean(abs(laplacian))
    vals = [max_val, min_val, median_val, mean_val]
    NCR_edges.append(vals)
    
# print(CR_edges[0:-1][0])
# print([i[0] for i in CR_edges])
# print(len(CR_edges))
```

```python
## histogram
hist1, bins1 = np.histogram([i[0] for i in CR_edges], bins=25)
width = 0.7 * (bins1[1] - bins1[0])
center = (bins1[:-1] + bins1[1:]) / 2

hist2, bins = np.histogram([i[0] for i in NCR_edges], bins=25)

plt.bar(center, hist2, align='center', width=width, label='Nope!')
plt.bar(center, hist1, align='center', width=width, alpha = .75, label = "Yeah!")
plt.title('Is it a cosmic ray? (By max edge change)')
plt.legend()
plt.show()
```

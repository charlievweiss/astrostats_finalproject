## Load data

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage.color import rgb2gray
```

```python
### load from data file

path = "processed_images/"
name = "data.txt"
filename = path + name

data = pd.read_csv(filename, delimiter = ",", engine='python')

#KATYA: Needed to define lists initially

images = []
img_nums = data.loc[:,'img_num']

### load images in grayscale

for i in range(0, len(data)):
    img_num = img_nums[i]
    img_file = "{}.png".format(img_num)
    img_name = path+img_file

    img = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)
    images.append(img)
    
# add images to dataframe for ease later
data['image'] = pd.Series(images, index=data.index)

### CHECK:
#img = data['image'].iloc[0]
#plt.imshow(img)
#print(data)
#print(labels)
#print(len(images))
```

## Create training and testing sets

```python
### split into training and testing sets
num_samples = data.shape[0]
training_size = .99 # percent of data in set (very large now for PCA later -- will generate more data for ML)
testing_size = 1 - training_size

# generate random sample from data for training set, and make testing set all else
msk = np.random.rand(num_samples) < training_size
training_data = data[msk]
testing_data = data[~msk]

### CHECK
#print(training_set)
#print(testing_set)
```

## Pre-process data with Principle Component Analysis
to extract meaningful features from data. Images are too big!

Resources:
- https://jakevdp.github.io/PythonDataScienceHandbook/05.09-principal-component-analysis.html

```python
# imports
from sklearn.decomposition import PCA
```

```python
### grab images from dataframe:
training_images = np.array(training_data.loc[:,'image'])

### Turn each image into vector for PCA
# 2d array: Array of image vectors
training_images_compressed = []

#Katya: What does image.flatten() do? Maybe comment for outside functions
# Katya: Comment what outside functions are doing 

for i,image in enumerate(training_images):
    temp = np.asarray(image.flatten())
    training_images_compressed.append(temp)
        
training_images_compressed = np.asarray(training_images_compressed)

### CHECK
#test = training_images[0]
#print(test.shape)
#print(training_images_compressed[0].shape)
print(training_images_compressed.shape)
```

```python
### apply PCA

#Katya: Why n_components = 2?
pca = PCA(n_components=2)
pca.fit(training_images_compressed)

### Reconstruct faces with principle components (for later figure)
components = pca.transform(training_images_compressed)
images_projected = pca.inverse_transform(components)

### CHECK
print(training_images_compressed.shape)
#print(pca.components_)
#print(pca.explained_variance_)
```

## Interesting plots from PCA

```python
### Plot first two principal commponets of each point to learn about the data
# 0 means not a cosmic ray, 1 means it is

## Get amount each image is represented by each component
plot_projected = pca.fit_transform(training_images_compressed)
## Get corresponding labels from dataframe
labels = np.asarray(training_data.loc[:,'label'])

## Plot
plt.scatter(plot_projected[:, 0], plot_projected[:, 1], c=labels,cmap='bwr');
plt.title("Images plotted with first 2 principle components")
plt.xlabel('component 1')
plt.ylabel('component 2')

# Legend
cbar = plt.colorbar(ticks=[0,1]);
cbar.ax.set_yticklabels(['not cosmic ray','cosmic ray']);

# Why would component 1 have more pictures with cosmic ray when its less than 0?
```

```python
### See how much info each projection preserves
# This is more informative if you apply PCA with max amount of components earlier

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');

plt.title("Increase in accuracy from 1 (min) to 36 (max) principle components")
```

```python
### Visualize Eigen Cosmic Rays
fig, axes = plt.subplots(1,2, figsize=(9, 4),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(pca.components_[i].reshape(6,6), cmap='bone')

fig.suptitle("Eigen Cosmic Rays (First 2 Components)");
```

```python
### Plot comparison between some original and reconstructions
fig, ax = plt.subplots(2, 10, figsize=(10, 2.5),
                       subplot_kw={'xticks':[], 'yticks':[]},
                       gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i in range(10):
    ax[0, i].imshow(training_images[i], cmap='binary_r')
    ax[1, i].imshow(images_projected[i].reshape(6,6), cmap='binary_r')
    
ax[0, 0].set_ylabel('full-dim\ninput')
ax[1, 0].set_ylabel('2-dim\nreconstruction');

fig.suptitle("Original vs Reconstructed Images");
```

## K-means clustering

```python
# imports
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering
from scipy.stats import mode
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# https://scikit-learn.org/stable/modules/clustering.html
# https://scikit-learn.org/stable/modules/classes.html#sklearn-metrics-metrics
```

Notes:

Given enough time, K-means will always converge, however this may be to a local minimum. This is highly dependent on the initialization of the centroids. As a result, the computation is often done several times, with different initializations of the centroids. One method to help address this issue is the k-means++ initialization scheme, which has been implemented in scikit-learn (use the init='k-means++' parameter). This initializes the centroids to be (generally) distant from each other, leading to provably better results than random initialization, as shown in the reference.


```python
# check datasets
# 535 images with 36 features (pixel brightness) each
training_images_compressed.shape
```

```python
# clustering
# result: 2 clusters in 36 dimensions. Each cluster is a 
# 36-dimensional point
kmeans = KMeans(n_clusters=2, random_state=0, n_init=50)
classifications = kmeans.fit_predict(training_images_compressed)
kmeans.cluster_centers_.shape
training_labels = np.array(training_data.loc[:,'label'])
# find accuracy
accuracy_score(training_labels, classifications)
```

```python
# check what cluster centers look like:
fig, ax = plt.subplots(2, 1, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(2, 6,6)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```
